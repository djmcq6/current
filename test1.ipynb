```python
ticker = 'BND'
data = yf.Ticker(ticker)
current_price = data.history(period='1d')['Close'].iloc[-1]

print(f"The current price of {ticker} is {current_price}")
```

```python
import yfinance as yf
import math

ticker = 'AAPL'
data = yf.Ticker(ticker)
current_price = data.history(period='1d')['Close'].iloc[-1]

rounded_price = math.ceil(current_price)
print(f"The current price of {ticker} is {rounded_price}")
```

```python
import yfinance as yf

ticker = 'TLT'
data = yf.Ticker(ticker)
current_price = data.history()

rounded_price = round(current_price, 2)
print(f"The current price of {ticker} is {rounded_price}")
```
